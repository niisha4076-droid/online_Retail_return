```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("retail_returns_data.csv")
```

```python
df['Returned'] = df['Returned'].str.lower()
```

```python
returned_df = df[df['Returned'] == 'yes']
```

```python
return_reason_count = returned_df['ReturnReason'].value_counts()
return_reason_count
```

```python
plt.figure(figsize=(8,5))
plt.bar(return_reason_count.index, return_reason_count.values)
plt.title("Product Return Reasons")
plt.xlabel("Return Reason")
plt.ylabel("Number of Returns")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
total_orders = len(df)
total_returns = len(returned_df)

print("Total Orders:", total_orders)
print("Returned Orders:", total_returns)
print("Return Percentage:", (total_returns / total_orders) * 100)
```
